```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings(action='ignore')
```

```python
data = pd.read_csv('final_data.csv', encoding='utf-8')
data.shape
```

```python
del data['Unnamed: 0']
data.dtypes
```

```python
data.head()
```

```python
group_df = data.groupby(['가맹점업종', '날짜']).sum().reset_index().set_index('가맹점업종')
group_df.head()
```

```python
upjong_list = data['가맹점업종'].unique().tolist()
print(upjong_list)
```

```python
# split multi dataframe based on 가맹점업종
subset_df = {}
for upjong in upjong_list:
    if upjong in group_df.index:
        subset_df[upjong] = group_df[group_df.index == upjong]
```

```python
subset_df.keys()
```

```python
health_retail = subset_df['건강보조식품 소매업']
health_retail.head()
```

```python
# 학습하려는 예측값 y값을 만들어주기
y_list = health_retail['카드사용금액'].iloc[1:].to_list()
y_list.insert(14, 0)
len(y_list)
```

```python
y_list
```

```python
health_retail['다음달_예측금액'] = y_list
health_retail.head()
```

```python
# y값인 '다음달 예측금액' 만들어주는 함수 정의
def create_y_value(df):
    y_list = df['카드사용금액'].iloc[1:].to_list()
    y_list.insert(len(y_list), 0)
    df['다음달_예측금액'] = y_list
    return df
```

```python
golf = subset_df['골프장 운영업']
golf_df = create_y_value(golf)
```
