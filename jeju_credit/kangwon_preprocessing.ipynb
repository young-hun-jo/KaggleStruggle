```python
upjong_list = data['가맹점업종'].unique().tolist()
print(upjong_list)
```

```python
# split multi dataframe based on 가맹점업종
subset_df = {}
for upjong in upjong_list:
    if upjong in group_df.index:
        subset_df[upjong] = group_df[group_df.index == upjong]
```

```python
# y값인 '다음달 예측금액' 만들어주는 함수 정의
def create_y_value(df):
    y_list = df['카드사용금액'].iloc[1:].to_list()
    y_list.insert(len(y_list), 0)
    df['다음달_예측금액'] = y_list
    return df
```

```python
# Linear Regression, Ride Regression
from sklearn.linear_model import LinearRegression, Ridge

features = health_retail.columns.tolist()[1:-1]

train_d, test_d = health_retail.iloc[:12], health_retail.iloc[12:-1]

train_y = train_d['다음달_예측금액']
train_x = train_d[features]

test_y = test_d['다음달_예측금액']
test_x = test_d[features]

models = [LinearRegression(), Ridge()]

for m in models:
    m.fit(train_x, train_y)
    pred = m.predict(test_x)
    score = m.score(test_x, test_y)
    print(f"모델 이름 : {m} , 정확도 : {score}, 예측값 : {pred}")
```

```python
from sklearn.neural_network import MLPRegressor

features = health_retail.columns.tolist()[1:-1]

train_d, test_d = health_retail.iloc[:12], health_retail.iloc[12:-1]

train_y = train_d['다음달_예측금액']
train_x = train_d[features]

test_y = test_d['다음달_예측금액']
test_x = test_d[features]

model = MLPRegressor()

model.fit(train_x, train_y)
score = model.score(test_x, test_y)
pred = model.predict(test_x)

print(f"모델 이름 : {model}, 정확도 : {score}, 예측값 : {pred}")
```

```python
data = pd.read_csv('encoded_data.csv', encoding='utf-8')
data.shape
```

```python
data.head()
```

```python
group_df = data.groupby(['날짜','가맹점주소','가맹점업종_이름']).sum().reset_index()
group_df
```

```python
group_df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

# 가맹점 주소 레이블 인코딩
region_encoder = encoder.fit_transform(group_df['가맹점주소'])
group_df['가맹점주소'] = region_encoder

# 업종별 레이블 인코딩
upjong_encoder = encoder.fit_transform(group_df['가맹점업종_이름'])
group_df['가맹점업종_이름'] = upjong_encoder

group_df.head()
```

```python
group_df.shape
```

```python
features = group_df.columns.tolist()
features.remove('카드사용금액')
features.remove('날짜')
print(features)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_log_error
train_d, test_d = group_df.iloc[:8829], group_df.iloc[8829:]

train_y = train_d['카드사용금액']
train_x = train_d[features]

test_y = test_d['카드사용금액']
test_x = test_d[features]

model = MLPRegressor()
model.fit(train_x, train_y)
acc = model.score(test_x, test_y)
pred = model.predict(test_x)
rmsle = mean_squared_log_error(test_y, pred)

# Regression에서 accuracy는 R2 score임
print(f"accuracy : {acc}, \n pred_value : {pred}")
print()
print(f" RMSLE : {rmsle}")
```
