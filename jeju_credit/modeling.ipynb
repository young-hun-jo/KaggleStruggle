```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings(action='ignore')
os.chdir('../../data')
```

```python
data = pd.read_csv('encoded_data.csv', encoding='utf-8')
data.head()
```

```python
data['가맹점주소'].nunique() * data['가맹점업종_이름'].nunique()
```

```python
cols = ['날짜', '가맹점주소', '가맹점업종_이름', '가맹점업종','사용자거주시도','연령대','가구생애주기','남자','여자']
group_df = data.groupby(cols).sum().reset_index()
```

```python
group_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# 가맹점주소 인코딩
encoder1= LabelEncoder()
sido_encoder = encoder1.fit_transform(group_df['가맹점주소'])
group_df['가맹점주소'] = sido_encoder

# 가맹점업종_이름 인코딩
encoder2 = LabelEncoder()
upjong_encoder = encoder2.fit_transform(group_df['가맹점업종_이름'])
group_df['가맹점업종_이름'] = upjong_encoder

group_df.head()
```

```python
final_data = group_df.groupby(['가맹점주소','가맹점업종_이름']).sum().reset_index()
final_data.head()
```

```python
del final_data['날짜']
```

```python
final_data.head()
```

```python
# 카드사용금액에 Log씌워주기
final_data['카드사용금액'] = np.log1p(final_data['카드사용금액'])
final_data.head()
```

```python
# feature normalization
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()

final_data[['사용자거주시도','연령대',
            '가구생애주기','남자','여자',
            '카드사용인원수(환불사용자포함)','카드사용건수(환불사용자제외)']] = scaler.fit_transform(final_data[['사용자거주시도','연령대',
                                                                                      '가구생애주기','남자','여자',
                                                                                      '카드사용인원수(환불사용자포함)','카드사용건수(환불사용자제외)']])
final_data.head()
```

```python
final_data.shape
```

```python
features = final_data.columns.tolist()[:-1]
features
```

```python
# 모델 학습시키기
from sklearn.neural_network import MLPRegressor

train_x = final_data[features]
train_y = final_data['카드사용금액']

model = MLPRegressor()
model.fit(train_x, train_y)
```

```python
temp = pd.DataFrame()
```

```python
temp = final_data[final_data.columns.tolist()[:-1]]
temp.head()
```

```python
pred = model.predict(temp)
pred = np.expm1(pred)
temp['4월예측값'] = pred
temp = temp[['가맹점주소','가맹점업종_이름','4월예측값']]
temp.head()
```

```python
temp['가맹점주소'] = encoder1.inverse_transform(temp['가맹점주소'])
temp['가맹점업종_이름'] = encoder2.inverse_transform(temp['가맹점업종_이름'])
temp.head(10)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_log_error
pca = PCA(n_components=6)

train_d, test_d = final_data.iloc[:585], final_data.iloc[585:]

train_y = train_d['카드사용금액']
train_x = pca.fit_transform(train_d[features])

test_y = test_d['카드사용금액']
test_x = pca.transform(test_d[features])

model = MLPRegressor()
model.fit(train_x, train_y)

score = model.score(test_x, test_y)
pred = model.predict(test_x)
msle = mean_squared_log_error(np.expm1(test_y), np.expm1(pred))

print(f"R2 score : {score}\n RMSLE : {msle} \n Predicted value : {np.expm1(pred)}")
```
